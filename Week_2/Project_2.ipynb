```python
import sympy as sp
x = sp.symbols('x')
print("Equation Calculator")
print("A. Quadratic Equation")
print("B. Cubic Equation")
print("C. Quartic Equation")
select = int(input("Select the type of equation from (1-3)"))
if select == 1:
    print("Quadratic eqn: ax^2+bx+c=0")
    a= float(input("Enter a: "))
    b= float(input("Enter b: "))
    c= float(input("Enter c: "))
    eqn = a*x**2+b*x+c
    roots = sp.solve(eqn, x)
    print("Roots are: ", roots)
elif select == 2:
    print("Cubic eqn: ax^3+bx^2+c*x+d=0")
    a= float(input("Enter a: "))
    b= float(input("Enter b: "))
    c= float(input("Enter c: "))
    d= float(input("Enter d: "))
    eqn = a*x**3+b*x**2+c*x+d
    roots = sp.solve(eqn, x)
    print("Roots are: ", roots)
elif select == 3:
    print("Quartic eqn: ax^4+bx^3+cx^2+d^x+e=0")
    a= float(input("Enter a: "))
    b= float(input("Enter b: "))
    c= float(input("Enter c: "))
    d= float(input("Enter d: "))
    e= float(input("Enter e: "))
    eqn = a*x**4+b*x**3+c*x**2+d
    roots = sp.solve(eqn, x)
    print("Roots are: ", roots)
else:
    print("Invalid Choice")

```
