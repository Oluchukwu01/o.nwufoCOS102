```python
print("Annual Tax Revenue Generator")
age = float(input("Enter your Age:"))
exp = int(input("Enter your years of work experience:"))

if exp >= 25 and age >= 55:
    print("Your Annual Tax Revenue, ATR is ₦5,600,000")

elif exp >= 20 and exp < 25 and age >= 45 and age < 55:
    print("Your Annual Tax Revenue, ATR is ₦4,480,000")
elif exp >= 10 and exp < 20 and age >= 35 and age < 45:
    print("Your Annual Tax Revenue, ATR is ₦1,500,000")
elif exp < 10 and age < 35:
    print("Your Annual Tax Revenue, ATR is ₦550,000")
else:
    print("Invalid input")
    
```
