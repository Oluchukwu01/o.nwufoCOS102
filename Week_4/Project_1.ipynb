```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

data = pd.read_csv("C:/Users/New/Downloads/archive/AB_NYC_2019.csv")
data
```

```python
data.isna().sum()
```

```python
data['last_review'] = pd.to_datetime(data['last_review'], errors='coerce')
```

```python
data.isna().sum()
```

```python
data['last_review'] = data['last_review'].fillna(data['last_review'].median())
```

```python
data.isna().sum()
```

```python
data['reviews_per_month'] = data['reviews_per_month'].fillna(data['reviews_per_month'].mean())
```

```python
data.isna().sum()
```

```python
data.dropna(inplace = True)
data
```

```python
data.isna().sum()
```

```python

data['price'].plot(kind='hist', bins=30)

plt.title('Distribution of Prices')
plt.xlabel('Price')
plt.ylabel('Frequency')
data['price'].plot(kind='hist', bins=50)

plt.xlim(0, 1500) 
plt.show()
```

```python
numeric_data = data.select_dtypes(include=['number'])
corr = numeric_data.corr()
```

```python
plt.figure(figsize=(10,8))
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
top = data['neighbourhood'].value_counts().head(10)

top.plot(kind='bar', figsize=(10,5))
plt.title('Top 10 Neighborhoods by Listings')
plt.show()
```
