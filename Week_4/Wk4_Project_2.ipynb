```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

data = pd.read_csv("C:/Users/New/Downloads/COS102/titanic.csv")
data
```

```python
data.isna().sum()
```

```python
data['Age'] = data['Age'].fillna(data['Age'].mean())
```

```python
data.isna().sum()
```

```python
data['Cabin'].isna().sum()
data['Cabin'].head()
```

```python
data.drop(columns=['Cabin'], inplace=True)
```

```python
data.isna().sum()
```

```python
data.dropna(inplace = True)
data
```

```python
data['Survived'].plot(kind='hist', bins=30)

plt.title('Survival Rate')
plt.xlabel('Survived')
plt.ylabel('Frequency')
data['Survived'].plot(kind='hist', bins=3)

plt.xlim(-1, 6) 
plt.show()
```

This shows that the rate of the people didn't survive is graeter than the rate of the survivors.

```python
data['Sex'] = data['Sex'].map({'male': 0, 'female': 1})
```

```python
numeric_data = data.select_dtypes(include=['number'])
corr = numeric_data.corr()
```

```python
plt.figure(figsize=(9,8))
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

There is a strong positive correlation between sex and survived.
There is a moderate negative correlation between Pclass and survived.
There is a very weak negative correlation between age and survived.
There is a weak negative correlation between Sibsp and survived.
There is a very weak positive correlation between Parch and survived.
There is a weak positive correlation between Fare and survived.
